```python
import lightkurve as lk
import numpy as np
import matplotlib.pyplot as plt
```

```python
searchResult = lk.search_lightcurve('KIC 12557548', author = 'Kepler')
print(searchResult)
```

```python
lcCollection = searchResult[2:6].download_all()
fig, ax = plt.subplots(figsize = (20,5))
lcCollection.plot(ax=ax);
```

```python
transitValues = []
for j in range(len(lcCollection)):
    lc = lcCollection[j]
    lc = lc.flatten()
    lc = lc.normalize()
    lc = lc.remove_nans()
    flux = lc.flux.value
    for i in range(len(flux)):
        if flux[i] < 0.9975:
            transitValues.append(1-flux[i])
```

```python
fig1, ax1 = plt.subplots()
x = np.ones(len(transitValues))
data = np.array(transitValues)
ax1.hist(data,50)
plt.show()
```
